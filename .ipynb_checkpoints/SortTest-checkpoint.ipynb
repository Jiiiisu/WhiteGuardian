```python
import os
from datetime import datetime, timedelta
from datetime import date

folder_path = os.path.join(os.getenv('APPDATA'), 'Microsoft', 'Windows', 'Recent')
file_list = os.listdir(folder_path)
today = date.today()
two_weeks_ago = today - timedelta(days=14)

# 해당 경로의 파일 이름 한줄씩 출력
#for f_name in recent_file_list:
#    print("%s"%f_name)
#print("\n")

#파일 목록을 수정 시간 기준으로 정렬 (내림차순)
sorted_files =  sorted(file_list, key=lambda x: os.path.getmtime(os.path.join(folder_path, x)), reverse=True)

#정렬된 파일 목록 출력. 이 값들 딕셔너리 형태로 저장. 
#for f_name in sorted_files:
#   file_path = os.path.join(folder_path,f_name)
#   time=os.path.getmtime(file_path)
#   date=datetime.fromtimestamp(time)
#   f_dic={f'{f_name}: {date}'}
#   print(f_dic)

#2주간 사용한 파일 목록 딕셔너리 형태로 저장.
recent_files = {}
for f_name in sorted_files:
    file_path = os.path.join(folder_path,f_name)
    time=os.path.getmtime(file_path)
    date=datetime.fromtimestamp(time).date()
    if two_weeks_ago <= date <= today:
        recent_files[f_name] = date

# 딕셔너리 형태로 저장된 최근 2주간 사용한 파일 목록을 출력
for f_name, date in recent_files.items():
    print(f'{f_name}: {date}일')
```
